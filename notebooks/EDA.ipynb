```python
import pandas as pd

df = pd.read_csv("../data/dataset.csv")

print(df.info())
print(df.isnull().sum())
print(df.describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="Lead_Converted", data=df)
plt.title("Lead Conversion Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Lead_Source", hue="Lead_Converted", data=df)
plt.title("Lead Conversion by Source")
plt.xticks(rotation=30)
plt.show()
```

```python
sns.histplot(df["Performance_Score"], bins=30, kde=True)
plt.title("Performance Score Distribution")
plt.show()
```

```python
pip_top = df[["PIP_Flag", "Top_Performer_Flag"]].sum()

pip_top.plot(kind="bar")
plt.title("PIP vs Top Performer Count")
plt.show()
```

```python
sns.countplot(x="Employment_Type", data=df)
plt.title("Employment Type Distribution")
plt.show()
```

```python
sns.boxplot(x="Employment_Type", y="CTC", data=df)
plt.title("CTC Distribution by Employment Type")
plt.show()
```
